```python
x = df[["YearsExperience"]]
y = df["Salary"]
```

```python
X_train, X_test, y_train, y_test =  train_test_split(x , y , test_size=0.2, random_state=2)
```

```python
model = LinearRegression()

 
```

```python
model.fit(X_train,y_train)
```

```python
model.predict([[8.32]])
```

```python
model.score(X_test,y_test)
```
